```python
#-*- coding:utf-8 -*-
import re #정규 표현식 모듈
import random
import math
```

```python
fp=open('8CT_0044.txt', 'r', encoding='utf-16')
ff=open('chat.log','w',encoding='utf-8')
```

```python
#Nstr = re.compile("\/[\W].*?")

cnt =0;
while True:
    line = fp.readline()

    if not line:
        break
        
    dialog = re.search('8CT_0044-[0-9]{7}',line,re.I|re.S)
    if (dialog is None): #대화문이 아닐 경우 반복문의 다음차례로 넘어감
        continue
    
    line = re.sub('8CT_0044-[0-9]{7}\t','',line ,0, re.I|re.S) 

    person = re.search('<u who="P\d">',line, re.I|re.S)
    if (person is not None):
        who = re.sub('[^0-9]','',line)
        print ("P"+who+ ": ", end='')
        
    personEnd = re.search('</u>',line, re.I|re.S)
    
    line = re.sub('<.*?>','',line ,0, re.I|re.S)
    line = re.sub(',','',line ,0, re.I|re.S)
    line = re.sub('\n','',line ,0, re.I|re.S)
    line = re.sub('<trunc>.*</trunc>','',line ,0, re.I|re.S)
    
    #한 사람이 말 한 내용이 없으면(한숨 같은 효과음) 없애기
    #line = re.sub('  ',' ',line ,0, re.I|re.S)
    
    
    #" ".join(line.split())



    
    if (personEnd is not None): #한사람 말 끝나면 줄넘김
        print(line)
        ff.write(line+"\n")
        cnt=cnt+1
    elif (line == ''):
        continue
    else: #말 이어서 하는중이면 줄넘김 없음
        print(line, end='')
        print(" ", end='')
        ff.write(line)
        
    if cnt == 2:
        print("")
        cnt = 0

fp.close()
ff.close()
```
